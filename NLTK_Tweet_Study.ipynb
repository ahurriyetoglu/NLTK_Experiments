```python
# natural language toolkit for syllable counting
import nltk
from nltk.corpus import cmudict
 
ENGLISH_STOPWORDS = set(nltk.corpus.stopwords.words('english'))
NON_ENGLISH_STOPWORDS = set(nltk.corpus.stopwords.words()) - ENGLISH_STOPWORDS
 
def is_english(text):
    """Return True if text is probably English, False if text is probably not English
    """
    text = text.lower()
    words = set(nltk.wordpunct_tokenize(text))
    return len(words & ENGLISH_STOPWORDS) > len(words & NON_ENGLISH_STOPWORDS)
```

```python
# analyse vocabulary

input_file = '/Users/doug/Desktop/sentiment/UCLA.json'
fd = open(input_file)
count = 10
error_count = 30
total_count = 300000
vocab = {}
for tweet_json in fd:
    try:
        tweet = json.loads(tweet_json)
        tweet_text = tweet['tweet']['text']
        tokens = nltk.word_tokenize(clean(tweet_text))
        for word in tokens:
            if word in vocab:
                vocab[word]+=1
            else:
                vocab[word]=1
    except ValueError:
        print "Error: " + tweet_json
        error_count -= 1
    total_count -= 1
    if (total_count <= 0) or (count <= 0) or (error_count <= 0):
        break
print 'done'
fd.close()
wordlist = vocab.keys()
```

```python
from nltk.corpus import wordnet as wn
known_words = [word for word in wordlist if wn.synsets(word.lower())]
unknown_words = [word for word in wordlist if not wn.synsets(word.lower())]
print 'done'
```
