```python
import re
url_pattern = re.compile('((http|https|HTTP|HTTPS)://\S*?)(\s|\)|\]|$)')
#http://stackoverflow.com/questions/13729638/how-can-i-filter-emoji-characters-from-my-input-so-i-can-save-in-mysql-5-5
try:
    emojis = re.compile(u'[\U00010000-\U0010ffff]')
except re.error:
    emojis = re.compile(u'[\uD800-\uDBFF][\uDC00-\uDFFF]')

def clean_tweet(text):
    if type(text) == unicode:
        text = re.sub(emojis, '', text)
        text = text.encode('ascii', 'ignore')
    text = text.replace('&amp;', 'and')
    
    #Remove @Usernames 
    text = re.sub('@[^\s]+','',text)
    
    # strip URLs  
    m = re.search(url_pattern, text)  
    while m:
        text = text[:m.start(1)] + text[m.end(1):]
        m = re.search(url_pattern, text)
    return text

    #Remove HTML
    text = nltk.clean_html(text)
    
    return text

print clean_tweet("this &amp; that")
print clean_tweet("my house (http://www.redhat.com) high there http://www.redhat.com")
```

```python
ignore_char = '~`!@#$%^&*()_-+={[}]|\\:;\"\'<,>.?/'+'1234567890'
def clean_token(text, lcase=True):
    #Remove repeated letters more than 2 times and Haaaaaaa = haa
    #http://stackoverflow.com/questions/10072744/remove-repeating-characters-from-words
    text = re.sub(r'(.)\1+', r'\1\1', text)  
    text = text.translate(None, ignore_char)
    if lcase:
        text = text.lower()
    return text
print clean_token("Soooooooweeeeeee")
print clean_token("Soooooooweeeeeee/beeee", lcase=False)
```

```python
name_pattern = re.compile('\A([^a-zA-Z]+)')
def clean_name(text, cap=True):
    # strip numbers after a name 
    m = re.search(name_pattern, text)  
    while m:
        text = text[:m.start(1)] + text[m.end(1):]
        m = re.search(url_pattern, text)
    if cap:
        text = text.capitalize()
    return text
print clean_name("doug")
print clean_name("doug1957", cap=False)
```

```python
# analyse vocabulary
import collections

input_file = '/Users/doug/Desktop/sentiment/UCLA.json'
fd = open(input_file)
count = 10
error_count = 30
total_count = 300000
vocab = collections.defaultdict(int)
for tweet_json in fd:
    try:
        tweet = json.loads(tweet_json)
        tweet_text = tweet['tweet']['text']
        tokens = nltk.word_tokenize(clean_tweet(tweet_text))
        for word in tokens:
            vocab[word]+=1
    except ValueError:
        print "Error: " + tweet_json
fd.close()
wordlist = frozenset(vocab.keys())
print 'done'
```

```python
# Load vocabulary
def load_vocab(source="all"):
    vocab = set([])
    if source in ["nltk_words", "all"]:
        vocab = vocab | set(w.lower() for w in nltk.corpus.words.words())
    if source in ["wordsEn", "all"]:
        #source: http://www-01.sil.org/linguistics/wordlists/english/wordlist/wordsEn.txt
        input_file = '/Users/doug/Desktop/sentiment/wordsEn.txt'
        fd = open(input_file)
        vocab = vocab | set(str.strip(w.lower()) for w in fd)
        fd.close()
    return vocab

known_vocab = load_vocab()
print "vocab size: {}".format(len(vocab))
print 'done'
#print wordEn_vocab
```

```python
from nltk.corpus import wordnet as wn
def known(word):
    word = clean_token(word)
    return (word in known_vocab) or wn.synsets(word)

known_words = frozenset([word for word in wordlist if known(word)])
unknown_words = frozenset(wordlist - known_words)
remaining_unknown = frozenset(unknown_words)
print 'known: {}'.format(len(known_words))
print 'unknown: {}'.format(len(unknown_words))
print 'done'
```

```python
all_names = nltk.corpus.names.words()
known_names = frozenset([word for word in remaining_unknown if clean_name(word) in all_names])
remaining_unknown = unknown_words.difference(frozenset(known_names))
print len(known_names)
```

```python
non_alpha_words = [word for word in unknown_words if not word.isalpha()]
print 'non-alpha words: {}'.format(len(non_alpha_words))
remaining_unknown = unknown_words.difference(frozenset(non_alpha_words))
print 'remaining unknown words: {}'.format(len(remaining_unknown))
#print list(remaining_unknown)[:50]
```

```python
def filtered_vocab(filter, threshold, isname=False, clean=True):
    result = collections.defaultdict(int)
    for key, value in vocab.items():
        if key in filter:
            if clean:
                if isname:
                    result[clean_name(key)] += value
                else:
                    result[clean_token(key)] += value
            else:
                result[key] += value
    for key, value in result.items():
        if value < threshold:
            del result[key]
    return result
```

```python
def filtered_stopwords(wordlist):
    result = collections.defaultdict(int)
    stop_words = nltk.corpus.stopwords.words('english')
    for key, value in wordlist.items():
        if not key in stop_words:
            result[key] = value
    return result
```

```python
filtered_known = filtered_stopwords(filtered_vocab(known_words, 2))
filtered_remaining_unknown = filtered_vocab(remaining_unknown, 5, clean=False)
filtered_non_alpha = filtered_vocab(non_alpha_words, 5, clean=False)
filtered_case_sensitive = filtered_vocab(case_sensitive_words, 5)
filtered_names = filtered_vocab(known_names, 5, isname=True)
print 'done'
```

```python
from nltk.probability import FreqDist
def makeFreqDist(series):
    result = FreqDist()
    for key, value in series.items():
        result.inc(key, count=value)
    return result
```
